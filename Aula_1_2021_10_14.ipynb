```python
## Contando o número de homens e mulheres
# Mudando o rótulo das colunas
PNAD_COVID_112020['Sexo'] = PNAD_COVID_112020['A003'].apply(lambda i: "Homem" if i == 1 else "Mulher")

# Contando usando value_counts
PNAD_COVID_112020.Sexo.value_counts(dropna=False)
```

```python
# Todas as colunas
PNAD_COVID_112020.groupby("Sexo").count()
```

```python
## Usando value_counts para obter valores percentuais
PNAD_COVID_112020.Sexo.value_counts(normalize=True, dropna=True)
```

```python
# Temos que usar groupby!
# Contando em termos absolutos
PNAD_COVID_112020.groupby("Sexo")["V1032"].sum()
```

```python
# Termos percentuais
contagem_sexo = PNAD_COVID_112020.groupby("Sexo")["V1032"].sum()
percentual_sexo = contagem_sexo / PNAD_COVID_112020["V1032"].sum() *100
percentual_sexo
```

```python
## Vendo gênero por estado
# Sem pesos
PNAD_COVID_112020.groupby(["UF", "Sexo"])["Sexo"].count().unstack()
```

```python
# Com pesos
contagem_sexo_uf = PNAD_COVID_112020.groupby(["UF", "Sexo"])["V1032"].sum()
pesos_uf = PNAD_COVID_112020.groupby("UF")["V1032"].sum()
percentual_sexo_uf = (contagem_sexo_uf / pesos_uf).unstack()
percentual_sexo_uf
#pesos_uf
```
