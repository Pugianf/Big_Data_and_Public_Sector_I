## Alternativa (mais abrangente) ao value_counts: groupby

```python
## Contando o número de homens e mulheres
# Mudando o rótulo das colunas
PNAD_COVID_112020['Sexo'] = PNAD_COVID_112020['A003'].apply(lambda i: "Homem" if i == 1 else "Mulher")

# Contando usando value_counts
PNAD_COVID_112020.Sexo.value_counts(dropna=False)
```

```python
## Contando usando groupby
# Apenas valores da própria coluna
PNAD_COVID_112020.groupby("Sexo")["Sexo"].count()
```

```python
# Todas as colunas
PNAD_COVID_112020.groupby("Sexo").count()
```

```python
## Usando value_counts para obter valores percentuais
PNAD_COVID_112020.Sexo.value_counts(normalize=True, dropna=True)
```

E os pesos da PNAD?

```python
## Temos que usar groupby!
# Contando em termos absolutos
PNAD_COVID_112020.groupby("Sexo")["V1032"].sum()
```

```python
# Termos percentuais
contagem_sexo = PNAD_COVID_112020.groupby("Sexo")["V1032"].sum()
percentual_sexo = contagem_sexo / PNAD_COVID_112020["V1032"].sum()
percentual_sexo
```

```python
## Vendo gênero por estado
# Sem pesos
PNAD_COVID_112020.groupby(["UF", "Sexo"])["Sexo"].count().unstack()
```

```python
# Com pesos
contagem_sexo_uf = PNAD_COVID_112020.groupby(["UF", "Sexo"])["V1032"].sum()
pesos_uf = PNAD_COVID_112020.groupby("UF")["V1032"].sum()
percentual_sexo_uf = (contagem_sexo_uf / pesos_uf).unstack()
percentual_sexo_uf
```

## Médias de Educação

```python
## Remapeando
Valores_PNAD = ["SI", "Fund Inc", "Fund Comp", "Méd Inc", "Med Comp", "Sup Inc", "Sup Comp", "Pós"]
PNAD_COVID_112020["Educação"] = PNAD_COVID_112020["A005"].replace([1, 2, 3, 4, 5, 6, 7, 8], Valores_PNAD)
```

```python
## Contando por escolaridade
np.around(PNAD_COVID_112020['Educação'].value_counts(normalize=True, dropna=True)*100, 2)
```

```python
## Gênero e Educação
# Sem pesos e relativo (frente ao total do gênero)
contagem_sexo_educ = PNAD_COVID_112020.groupby(["Sexo", "Educação"])["Educação"].count()
total_sexo = PNAD_COVID_112020.groupby("Sexo")["Sexo"].count()
(contagem_sexo_educ / total_sexo).unstack()
```

```python
# Sem pesos e relativo (frente ao total do grau de educação)
contagem_educ_sexo = PNAD_COVID_112020.groupby(["Educação", "Sexo"])["Educação"].count()
total_educ = PNAD_COVID_112020.groupby("Educação")["Educação"].count()
(contagem_educ_sexo / total_educ).unstack()
```
